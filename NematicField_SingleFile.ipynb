```python
import numpy as np
import os 
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
fileDir = "/home/srj_bruh/Acads/dir0.dat"

with open(fileDir,"r") as f:
    dataFile = []
    data = f.readlines()
    for line in data:
        words = line.split()
        dataFile.append(float(words[3])*10**-5)
```

```python
nx = 400
ny = 400

x = np.linspace(0,nx,nx)
y = np.linspace(0,ny,ny)
```

```python
dataFile = np.array(dataFile)
dataFile = dataFile.reshape((ny,nx))

plt.contourf(x,y,dataFile)
plt.rcParams.update({'font.size': 22})
rc('xtick', labelsize=22)
rc('ytick', labelsize=22)
plt.colorbar()
```
